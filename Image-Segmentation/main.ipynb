```python
import matplotlib as mpl
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
image_1 = mpl.image.imread("canguru.jpeg")
image_2=mpl.image.imread("dog.jpeg")

plt.imshow(image_1)
```

```python
plt.imshow(image_2)
```

```python
print(image_1.shape)
print(image_2.shape)
```

```python
X=image_1.reshape(-1,3)
X_2=image_2.reshape(-1,3)
```

```python
kmeans_1=KMeans(n_clusters=2,n_init=10)
kmeans_2=KMeans(n_clusters=2,n_init=10)

kmeans_1.fit(X)
kmeans_2.fit(X_2)
```

```python
segmented_img=kmeans_1.cluster_centers_[kmeans_1.labels_]
segmented_img_2=kmeans_2.cluster_centers_[kmeans_2.labels_]

segmented_img=segmented_img.reshape(image_1.shape)
segmented_img_2=segmented_img_2.reshape(image_2.shape)
```

```python
plt.imshow(segmented_img/255)
```

```python
plt.imshow(segmented_img_2/255)
```
